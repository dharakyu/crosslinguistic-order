```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part 1: consolidate the Kinbank data

```python
languages_df = pd.read_csv('raw_corpus_data/kinbank/languages.csv')
languages_df = languages_df.rename({'ID': 'Language_ID'}, axis=1)
languages_df
```

```python
forms_df = pd.read_csv('raw_corpus_data/kinbank/forms.csv')
forms_df
```

```python
num_forms_df = pd.DataFrame(forms_df.groupby('Language_ID')['Value'].nunique())
num_forms_df
```

```python
language_and_num_forms_df = languages_df.merge(num_forms_df, on='Language_ID')
language_and_num_forms_df
```

## Part 2: consolidate the WALS data (and get the ISO-639 codes)

```python
order_df = pd.read_csv('~/Documents/word-ordering/wals/order_adj_noun.tsv', sep='\t')
order_df
```

```python
language_codes_df = pd.read_csv('~/Documents/word-ordering/wals/languages.csv')
language_codes_df
```

```python
# get the ISO codes for WALS data
WALS_iso_df = pd.merge(order_df, language_codes_df, left_on='wals code', right_on='ID')
WALS_iso_df
```

```python
WALS_iso_df[WALS_iso_df['wals code'] == 'deu']
```

## Part 3: combine Kinbank and WALS

```python
order_and_count_df = pd.merge(WALS_iso_df, language_and_num_forms_df, on='ISO639P3code')
order_and_count_df
```

```python
# get rid of the duplicates
order_and_count_df = order_and_count_df.drop_duplicates(subset=['ISO639P3code'])
order_and_count_df
```

```python
order_and_count_df[order_and_count_df['ISO639P3code'] == 'deu']
```

```python
order_and_count_df.groupby('description')['Value'].mean()
```
